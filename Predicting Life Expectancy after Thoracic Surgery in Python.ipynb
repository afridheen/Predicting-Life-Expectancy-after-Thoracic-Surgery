### Feature Selection
+ SelectKBest
+ + Select the features with the strongest relationship with output
+ Recursive Feature Elimination
+ + Identify which features contribute the most to predicting the target output

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
```

```python
df1.head()
```

```python
# Features 
Xfeatures = df1.values[:,0:15]
```

```python
Ylabel = df1['Risk1Y']
```

```python
print(Xfeatures.shape)
```

```python
# Using a model to check 
from sklearn.linear_model import LogisticRegression
# Don't be confuse about the name it is a classification algorithm not a regression
modelLogit = LogisticRegression()
skb = SelectKBest(score_func=chi2,k=4)
skbfit = skb.fit(Xfeatures,Ylabel)
```

```python
skbfit.scores_
best_features = skbfit.transform(Xfeatures)
```

```python
# Highest scores show the best features
print("Scores",skbfit.scores_)
```

```python
# summarize selected features
print(best_features[0:15,:])
```

```python
mybestscores = list(skbfit.scores_)
myfeaturenames = ['Diagnosis', 'FEV', 'Performance', 'Pain', 'Haemoptysis', 'Dyspnoea',
       'Cough', 'Weakness', 'TNM', 'DM', 'MI', 'PAD', 'Smoking', 'Asthma',
       'AGE']
```

```python
for b,f in zip(mybestscores,myfeaturenames):
    print(f'"Feature":{f} => {b*100}')
```

```python
# Hence the highest feature scores are FEV,DM,Dyspnea,TNM,Haemoptysis
```

#### Using Recursive Feature Elimination to check which feature contribute the most to predicting the target results

```python
from sklearn.feature_selection import RFE
rfe = RFE(modelLogit,3)
rfe_fit = rfe.fit(Xfeatures,Ylabel)
```

```python
print("Number of Features",rfe_fit.n_features_)
print("Feature Names",list(df1.columns)[0:15])
print("Selected Features",rfe_fit.support_)
print("Features Ranking",rfe_fit.ranking_)
```

```python
# Based on the RFE dyspnea,weakness,DM contribute 
# Interesting results
```

#### Feature Importance 
+ The larger the score the more important the feature or attribute
+ ExtraTreesClassifier

```python
from sklearn.ensemble import ExtraTreesClassifier
modelExtra = ExtraTreesClassifier()
modelExtra.fit(Xfeatures,Ylabel)
print(modelExtra.feature_importances_)
```

```python

myimportantfeature = list(modelExtra.feature_importances_)
```

```python
for imp,mf in zip(myimportantfeature,myfeaturenames):
    print(f'"Feature":{mf} => {imp*100}')
```

```python
# hence the higher the score the more important the feature
# Diagnosis,FEV,Age,TNM,Performance,Hemoptysis,Weakness
```

### Evaluating the ML Algorithm Performance
+ Split into train,test
+ Kfold into K number of folds
+ LeaveOneOut

```python
Xfeatures
Ylabel
```

```python
# Split Data set
from sklearn.model_selection import train_test_split
seed = 7 # used to enable randomness being reproducible
X_train,X_test,Y_train,Y_test = train_test_split(Xfeatures,Ylabel,test_size=0.33,random_state=7)
```

```python
#using our LogisticRegression
modelLogit.fit(X_train, Y_train)
result = modelLogit.score(X_test, Y_test)
print("Accuracy: ",result*100.0)
```

```python
# Hence we our model is ok 83.9 %
```

```python
# using Cross Validation with KFold
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
```

```python
# Using our LogisticRegression
num_folds = 10
seed = 7
kfold = KFold(n_splits=num_folds, random_state=seed)
results = cross_val_score(modelLogit, Xfeatures, Ylabel, cv=kfold)
print(results)
print("Mean Accuracy:", results.mean()*100.0)
print("STD Accuracy:", results.std()*100.0)
```

```python
# Compare Algorithms
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
# load dataset

# prepare models
models = []
models.append(('LR', LogisticRegression()))
models.append(('LDA', LinearDiscriminantAnalysis()))
models.append(('KNN', KNeighborsClassifier()))
models.append(('CART', DecisionTreeClassifier()))
models.append(('NB', GaussianNB()))
models.append(('SVM', SVC()))
# evaluate each model in turn
results = []
names = []
scoring = 'accuracy'
for name, model in models:
    kfold = KFold(n_splits=10, random_state=7)
    cv_results = cross_val_score(model, Xfeatures, Ylabel, cv=kfold, scoring=scoring)
    results.append(cv_results)
    names.append(name)
    msg = "%s: %f (%f)" % (name, cv_results.mean(), cv_results.std())
```

```python

print(msg)
# boxplot algorithm comparison
fig = pyplot.figure()
fig.suptitle('Algorithm Comparison')
ax = fig.add_subplot(111)
pyplot.boxplot(results)
ax.set_xticklabels(names)
pyplot.show()
```
